```python
d = ntu.groupby('Entity')['No. of Internet Users'].mean().round().nlargest(7).reset_index()
sns.barplot(data=d, x = 'Entity', y = 'No. of Internet Users' )
```

```python
d = d.drop(d[(d['Entity']=='World')].index)
d = d.drop(d[(d['Entity']=='North America')].index)
d
```

```python
fig,ax = plt.subplots(figsize=(15,9))
sns.barplot(data=d, x = 'Entity', y = 'No. of Internet Users' )
ax.bar_label(ax.containers[0])
```

Top 10 countries with highest avrage user in 2020.

```python
u = ntu[ntu['Year']==2020].groupby('Entity')['No. of Internet Users'].mean().round().nlargest(10).reset_index()
u = u.drop(u[(u['Entity']=='world')].index)
u = u.drop(u[(u['Entity']=='North America')].index)
```

```python
fig, ax = plt.subplots(figsize = (15,9))
sns.barplot(data = u, x = 'Entity', y = 'No. of Internet Users')
ax.bar_label(ax.containers[0])
```
